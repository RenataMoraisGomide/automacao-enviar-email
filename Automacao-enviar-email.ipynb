### Passo a passo:

- Passo 2: Realizar as análises solicitadas pelo gestor/chefe
- Passo 3: Enviar automaticamente o e-mail com o resultado da análise.

# Passo 1


```python
 #Aqui o python esta dizendo que foi lá no site yahoo.finance e 
 #achou o objeto desejado(PETR4.SA)
 #(d)= dia (y)= ano (mo)= meses

import yfinance as yf

ticker = input('Digite o código da ação: ')
dados = yf.Ticker(ticker).history('6mo')        
fechamento = dados.Close
fechamento.plot()

```

# Passo 2: realizar as análise solicitadas pelo chefe
- cotação máxima
- cotação mínima
- cotação atual

```python
fechamento
```

```python
maxima = round(fechamento.max(),2)
minima = round(fechamento.min(),2)
atual = round(fechamento[-1],2)
print(maxima)
print(minima)
print(atual)
```

# Passo 3: Enviar automaticamente o e-mail com o resultado da análise.
- Abrir uma nova aba (ctrl + t)
- Digitar www.gmail.com e dar enter
- clicar no botão escrever
- Digitar o email do destinatário (tab)
- Escrever assunto (tab)
- Preencher o corpo do e-mail
- Clicar no botão enviar

```python
import pyautogui
import pyperclip
import time
```

```python
#Abrir uma nova aba (ctrl + t)
#Colocar uma pequena pausa para não bugar se estiver lento

pyautogui.PAUSE = 3
pyautogui.hotkey('ctrl', 't')

#Digitar www.gmail.com e dar enter
pyperclip.copy('www.gmail.com')
pyautogui.hotkey('ctrl', 'v')
pyautogui.hotkey('enter')


#clicar no botão escrever

#verificar a coordenada dos eixos de pixels usando o tempo do import time
#import time
#time.sleep(5)
#pyautogui.position()


pyautogui.click(x=90, y=205)

#Digitar o email do destinatário (tab)

pyperclip.copy('decorarbarbante@gmail.com')
pyautogui.hotkey('ctrl', 'v')
pyautogui.hotkey('tab')

#digitar o assunto:

pyperclip.copy('Análises diárias')
pyautogui.hotkey('ctrl', 'v')
pyautogui.hotkey('tab')


#Preencher o corpo do email:

mensagem = f'''Prezado Gestor, 
Seguem, às análises dos últimos seis meses 
conforme solicitado. {ticker}

Cotação Máxima: R$ {maxima}
Cotação Mínima: R$ {minima}
Cotação Atual: R$ {atual}


Qualquer dúvida, estou à disposição.
Atte.
Renata Morais.
'''

pyperclip.copy(mensagem)
pyautogui.hotkey('ctrl','v')


#cliclar no enviar:
#import time
#time.sleep(5)
#pyautogui.position()


pyautogui.click(x=840, y=695)

print('email enviado com sucesso')
```
